```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
dummy = pd.get_dummies(df['diagnosis'])
```

```python
df = pd.concat([df,dummy],axis=1)
```

```python
df
```

```python
df = df.drop(['id','diagnosis','M','Unnamed: 32'],axis=1)
df.head()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='B',data=df)
```

```python
df.describe().transpose()
```

```python
plt.figure(figsize=(12,8))
df.corr()['B'].sort_values().plot(kind = 'bar')
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr())
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('B',axis=1).values
y = df['B'].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout

model = Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),verbose=1,callbacks=[early_stop],epochs=600)
```

```python
losses = pd.DataFrame(model.history.history)
losses.plot()
```

```python
predictions = model.predict_classes(X_test)
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
